# C) Model Building:
## a. Build a sequential model using Keras, to find out if the customer would churn or not, using ‘tenure’ as the feature and ‘Churn’ as the dependent/target column:
### i. The visible/input layer should have 12 nodes with ‘Relu’ as activation function.
### ii. This model would have 1 hidden layer with 8 nodes and ‘Relu’ as activation function
### iii. Use ‘Adam’ as the optimization algorithm
### iv. Fit the model on the train set, with number of epochs to be 150
### v. Predict the values on the test set and build a confusion matrix
### vi. Plot the ‘Accuracy vs Epochs’ graph

```python
from keras.models import Sequential
from keras.layers import Input, Dense, Dropout
from keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
X=data["tenure"]
y=data["Churn"].map({'Yes': 1, 'No': 0})
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
model_a=Sequential()
```

```python
model_a.add(Dense(12, input_dim=1, activation='relu')) #Input layer
```

```python
model_a.add(Dense(8, activation='relu')) #Hidden layer
```

```python
model_a.add(Dense(1, activation='sigmoid')) #Output Layer
```

```python
model_a.compile(optimizer=Adam(learning_rate=0.002),loss="binary_crossentropy", metrics=["accuracy"])
```

```python
model_a.fit(X_train, y_train, epochs=150, validation_data=(X_test, y_test), batch_size=120)
```

```python
train_acc = model_a.history.history["accuracy"]
val_acc = model_a.history.history["val_accuracy"]

plt.plot(train_acc)
plt.plot(val_acc)
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend(["Train", "Test"])
plt.title("Accuracy vs Epochs")
plt.grid()
plt.show()
```

```python
y_pred=model_a.predict_classes(X_test)
conf=confusion_matrix(y_test,y_pred)
```

```python
print("Confusion matrix:\n")
print(conf)
```

```python
print("Mean validation accuracy for the model:",np.mean(val_acc))
```
